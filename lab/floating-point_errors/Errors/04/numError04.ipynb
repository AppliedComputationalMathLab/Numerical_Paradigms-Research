````python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import plotly.express as px
import pandas as pd 
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
import plotly.figure_factory as ff
import plotly.io as pio
import plotly.subplots as sp
import plotly.express as px
````

````python
def f(x: np.float64) -> np.float64:
    return ((1+x)-1) / x
````

````python
rangeX = (-4e-15, 4e-15)
x = np.linspace(rangeX[0], rangeX[1], 20000, dtype=np.float64)
y = f(x)
````

````python
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42

plt.plot(x, y, label='g(x) = ((1+x)-1) / x', color='red', linewidth=1.3)
plt.xlabel('x')
plt.ylabel('g(x)= ((1+x)-1) / x')
plt.gcf().set_size_inches(18, 6)
plt.legend(fontsize=15)
plt.xlim(rangeX)


centimeters_x = 27
centimeters_y = 11.5

inches_x = centimeters_x / 2.54
inches_y = centimeters_y / 2.54


plt.gcf().set_size_inches(inches_x, inches_y)

plt.grid(True, linewidth=1.0, alpha=1)

plt.savefig(
    "graphs/poster/NumError4_mplt.pdf",
    format='pdf',
    bbox_inches='tight',
)
````
