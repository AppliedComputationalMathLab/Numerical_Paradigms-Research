```python
import numpy as np
import matplotlib.pyplot as plt


def mandelbrot_orbit(c, n_iter):
    z = 0+0j
    orbit = []
    for _ in range(n_iter):
        z = z*z + c
        orbit.append(z)
    return np.array(orbit)


# ponto próximo da borda do Mandelbrot (região sensível)
c = np.complex128(-0.743643887037151 + 0.131825904205330j)

# cria c' diferindo de c por 1 ULP
c_ulp = np.nextafter(c.real, np.inf) + 1j*np.nextafter(c.imag, np.inf)

# número de iterações
N = 200

orbit1 = mandelbrot_orbit(c, N)
orbit2 = mandelbrot_orbit(c_ulp, N)

diff = np.abs(orbit1 - orbit2)

plt.figure(figsize=(8,5))
plt.plot(diff)
plt.yscale("log")
plt.xlabel("Iteração n")
plt.ylabel("|z_n - z'_n|")
plt.title("Amplificação de diferença de 1 ULP na órbita do Mandelbrot")
plt.grid(True)

plt.show()
```
