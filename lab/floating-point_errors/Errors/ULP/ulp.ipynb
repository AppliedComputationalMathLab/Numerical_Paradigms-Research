```python
import plotly.graph_objects as go

def generate_float_scale(min_power, max_power):
    """Gera uma lista de pontos representando números de ponto flutuante, 
       mostrando o aumento do ULP (espaçamento)."""
    
    points = []
    
    # 1. Pontos no intervalo [2^(min_power), 2^(min_power + 1)]
    # ULP = 2^(min_power - 52 + 1) -> vamos usar 2^min_power
    start = 2**min_power
    ulp_small = 0.5 * start 
    for i in range(5): # 5 marcas
        points.append(start + i * ulp_small)

    # 2. Pontos no intervalo [2^(min_power + 1), 2^(min_power + 2)]
    # ULP dobra a partir da potencia de 2
    start = 2**(min_power + 1) 
    ulp_large = 0.5 * start # ULP dobrou
    for i in range(5): # 5 marcas
        points.append(start + i * ulp_large)
        
    return sorted(list(set(points)))

# Definindo as potências para a visualização
# Vamos visualizar a transição de expoente de 2^1 para 2^2
min_pow = 1
max_pow = 2

float_points = generate_float_scale(min_pow, max_pow)

# --- Criação do Gráfico no Plotly ---

fig = go.Figure()

# 1. Plotar a linha horizontal
fig.add_trace(go.Scatter(
    x=[min(float_points) - 0.2, max(float_points) + 0.2],
    y=[0, 0],
    mode='lines',
    line=dict(color='black', width=1),
    name='Eixo Numérico',
    showlegend=False
))

# 2. Plotar as marcas dos pontos flutuantes (Discretização)
for point in float_points:
    fig.add_trace(go.Scatter(
        x=[point, point],
        y=[-0.05, 0.05], # Desenha uma pequena linha vertical
        mode='lines',
        line=dict(color='black', width=1.5),
        showlegend=False
    ))

# 3. Adicionar o Rótulo da Transição (Potência de 2)
# Vamos destacar o ponto 2^2 = 4
transition_point = 2**max_pow 
fig.add_annotation(
    x=transition_point,
    y=-0.25,
    text=f"$2^{{{max_pow}}}$", # Texto com LaTeX para 2^j (j=2)
    showarrow=False,
    font=dict(size=14)
)

# 4. Simular o Arredondamento (O "Erro")
# Ponto Exato (real) e Ponto Calculado (arredondado)
exact_value = transition_point + 0.3 # Um valor que precisa ser arredondado
computed_value = transition_point + 0.5 # Arredonda para o ponto representável mais próximo

# Adicionar o valor exato (real)
fig.add_trace(go.Scatter(
    x=[exact_value],
    y=[0],
    mode='markers',
    marker=dict(size=8, color='blue'),
    name='Valor Exato (Real)',
    hoverinfo='text',
    text=[f'Exato: {exact_value}']
))
fig.add_annotation(
    x=exact_value,
    y=0.15,
    text="Exato",
    showarrow=True,
    arrowhead=1,
    arrowsize=1,
    arrowwidth=1,
    ax=0,
    ay=-30
)

# Adicionar o valor calculado (arredondado)
fig.add_trace(go.Scatter(
    x=[computed_value],
    y=[0],
    mode='markers',
    marker=dict(size=8, color='red'),
    name='Valor Arredondado (Computado)',
    hoverinfo='text',
    text=[f'Calculado: {computed_value}']
))
fig.add_annotation(
    x=computed_value,
    y=0.15,
    text="Calculado",
    showarrow=True,
    arrowhead=1,
    arrowsize=1,
    arrowwidth=1,
    ax=0,
    ay=-30,
    yshift=-20 # Desloca ligeiramente o rótulo
)

# 5. Configuração do Layout
fig.update_layout(
    title='Visualização da Discretização de Ponto Flutuante e Crescimento do ULP',
    xaxis_title='Valores',
    yaxis_title='',
    yaxis=dict(showticklabels=False, range=[-0.5, 0.5]),
    xaxis=dict(zeroline=False, showgrid=False, tickvals=float_points, ticktext=['' for _ in float_points]),
    showlegend=True
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

def generate_float_density(min_power=-5, max_power=5, points_per_power=5):
    """Gera pontos simulando densidade de números de ponto flutuante."""
    points = []
    for exp in range(min_power, max_power + 1):
        start = 2**exp
        ulp = start / points_per_power
        for i in range(points_per_power):
            points.append(start + i * ulp)
        # também adiciona negativos
        for i in range(points_per_power):
            points.append(-start - i * ulp)
    return sorted(points)

float_points = generate_float_density()

# --- Criação do Gráfico ---
fig = go.Figure()

# Linha horizontal
fig.add_trace(go.Scatter(
    x=[min(float_points)*1.1, max(float_points)*1.1],
    y=[0,0],
    mode='lines',
    line=dict(color='black', width=1)
))

# Marcas dos pontos
for point in float_points:
    fig.add_trace(go.Scatter(
        x=[point, point],
        y=[-0.05,0.05],
        mode='lines',
        line=dict(color='black', width=1)
    ))

# Configuração dos eixos
fig.update_layout(
    title='Escala de Ponto Flutuante: Densidade Próxima a Zero',
    xaxis=dict(
        title='Valores',
        zeroline=True,
        showgrid=True
    ),
    yaxis=dict(
        title='',
        showticklabels=False,
        zeroline=True
    ),
    showlegend=False,
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

def generate_float_density(min_power=-5, max_power=5, points_per_power=5):
    """Gera pontos simulando densidade de números de ponto flutuante."""
    points = []
    for exp in range(min_power, max_power + 1):
        start = 2**exp
        ulp = start / points_per_power
        for i in range(points_per_power):
            points.append(start + i * ulp)
        # também adiciona negativos
        for i in range(points_per_power):
            points.append(-start - i * ulp)
    return sorted(points)

float_points = generate_float_density()

tick_height = 1e-12

fig = go.Figure()

# Eixo horizontal
fig.add_trace(go.Scatter(
    x=[min(float_points)*1.1, max(float_points)*1.1],
    y=[0,0],
    mode='lines',
    line=dict(color='black', width=1),
    showlegend=False
))

# Marcas dos pontos representáveis
for point in float_points:
    fig.add_trace(go.Scatter(
        x=[point, point],
        y=[-tick_height, tick_height],
        mode='lines',
        line=dict(color='gray', width=1),
        showlegend=False
    ))

# Valores escolhidos
exact_value = 30
computed_value = 28.8

# Encontrar os dois floats mais próximos do valor exato
nearest_points = sorted(float_points, key=lambda x: abs(x - exact_value))[:2]
left, right = sorted(nearest_points)

# Destacar a faixa de arredondamento
fig.add_shape(
    type="rect",
    x0=left,
    x1=right,
    y0=-tick_height*10,
    y1=tick_height*10,
    fillcolor="rgba(255, 200, 200, 0.4)",
    line=dict(width=0),
    layer="below"
)

# Ponto exato (valor real)
fig.add_trace(go.Scatter(
    x=[exact_value],
    y=[0],
    mode='markers+text',
    marker=dict(size=10, color='blue'),
    text=['exato'],
    textposition='top center'
))

# Ponto computado (arredondado)
fig.add_trace(go.Scatter(
    x=[computed_value],
    y=[0],
    mode='markers+text',
    marker=dict(size=10, color='red'),
    text=['computado'],
    textposition='bottom center'
))

# Seta de arredondamento
fig.add_annotation(
    x=exact_value,
    y=0,
    ax=computed_value,
    ay=0,
    arrowhead=3,
    arrowsize=1,
    arrowwidth=1.5,
    arrowcolor='red',
    text="arredondamento",
    font=dict(size=12),
    yshift=30
)

# Layout
fig.update_layout(
    title='Demonstração do Arredondamento entre Pontos de Ponto Flutuante',
    xaxis=dict(
        title='Valores',
        zeroline=True,
        showgrid=True
    ),
    yaxis=dict(
        showticklabels=False,
        range=[-tick_height * 20, tick_height * 20],
        zeroline=False
    ),
    showlegend=False,
    height=520
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# --- Parâmetros do sistema simplificado ---
b = 2
p = 4
e = 2
quantum = b ** (e - p + 1)          # 2^(e - p + 1)
ulp_harrison = b ** (e - p)         # 2^(e - p) = 0.25

# --- Números representáveis ---
start = 2 ** e
representaveis = np.arange(start, 2 ** (e + 1) + quantum, quantum)

# --- Valor real e arredondado ---
x_real = 4.3
x_float = 4.5
erro = abs(x_real - x_float)
erro_quantum = erro / quantum
erro_ulp_h = erro / ulp_harrison

# --- Gráfico base ---
fig = go.Figure()

# Linha horizontal principal
fig.add_trace(go.Scatter(
    x=[start - 0.2, 2 ** (e + 1) + 0.2],
    y=[0, 0],
    mode='lines',
    line=dict(color='black', width=2),
    showlegend=False
))

# Marcas verticais (números representáveis)
for pt in representaveis:
    fig.add_trace(go.Scatter(
        x=[pt, pt],
        y=[-0.05, 0.05],
        mode='lines',
        line=dict(color='gray', width=2),
        showlegend=False
    ))

# Valor real (azul)
fig.add_trace(go.Scatter(
    x=[x_real],
    y=[0],
    mode='markers+text',
    marker=dict(size=12, color='blue'),
    text=["valor real"],
    textposition='top center',
    name='Valor real'
))

# Valor arredondado (vermelho)
fig.add_trace(go.Scatter(
    x=[x_float],
    y=[0],
    mode='markers+text',
    marker=dict(size=12, color='red'),
    text=["ponto flutuante (arredondado)"],
    textposition='bottom center',
    name='Valor arredondado'
))

# Seta de erro
fig.add_annotation(
    x=x_real,
    y=0.12,
    ax=x_float,
    ay=0.12,
    text=f"erro = {erro:.2f}\n({erro_quantum:.1f} Quantum) -\n ({erro_ulp_h:.1f} ULPₕ)",
    showarrow=True,
    arrowhead=2,
    arrowsize=1.5,
    arrowwidth=1.5,
    arrowcolor='orange',
    font=dict(size=14, color='orange'),
    align='left'
)

# Intervalo de arredondamento (opcional)
fig.add_trace(go.Scatter(
    x=[4.0, 4.5],
    y=[-0.15, -0.15],
    mode='lines',
    line=dict(color='orange', width=3, dash='dot'),
    showlegend=False
))

# --- Layout ---
fig.update_layout(
    title='Erro de Arredondamento: Quantum vs ULP de Harrison',
    xaxis=dict(
        title='Valores representáveis no intervalo [2², 2³)',
        showgrid=False,
        zeroline=False,
        tickmode='array',
        tickvals=representaveis,
        ticktext=[f"{v:.2f}" for v in representaveis]
    ),
    yaxis=dict(showticklabels=False, range=[-0.3, 0.3]),
    showlegend=False,
    template='plotly_white',
    height=450
)

fig.show()
```
