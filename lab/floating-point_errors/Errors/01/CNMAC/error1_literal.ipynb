````python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib as mp
import matplotlib.ticker as ticker
import seaborn as sns
import plotly.express as px
import pandas as pd 
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
import plotly.figure_factory as ff
import plotly.io as pio
import plotly.subplots as sp
import plotly.express as px
````

````python
def f(x: np.float64, literal: np.float64) -> np.float64:
    return (x**23) + literal - (x**23)
````

````python
def getULP(x: np.ndarray) -> np.ndarray:
    x = np.asarray(x, dtype=np.float64)
    return np.nextafter(x, np.inf) - x
````

````python
mp.rcParams['pdf.fonttype'] = 42
mp.rcParams['ps.fonttype'] = 42
````

````python
centimeters_x = 27
centimeters_y = 11.5

inches_x = centimeters_x / 2.54
inches_y = centimeters_y / 2.54


plt.gcf().set_size_inches(inches_x, inches_y)
````

````python
rangeX = (4.9, 6)
````

````python
fig = plt.figure()
plt.xlabel('x')
plt.ylabel('f(x)')

plt.gcf().set_size_inches(18, 6)

x = np.linspace(rangeX[0], rangeX[1], 2200, dtype = np.float64)

for i in np.arange(1, 32,2, dtype=np.float64):
    y = f(x, i)
    plt.plot(x, y, label=f'Literal = {i}')


plt.legend(fontsize=12)
plt.xlim(rangeX)

plt.grid(True, linewidth=1.0, alpha=1)

plt.legend(loc=1, fontsize=10)


fig.show()
````

````python
x = np.linspace(rangeX[0], rangeX[1], 1200, dtype=np.float64)

fig2, (ax1, ax2) = plt.subplots(2, 1, figsize=(inches_x*2, inches_y*2), sharex=True)

q = 6

colors = plt.cm.viridis(np.linspace(0, 1, q))

for idx, n in enumerate(np.arange(0, q)):
    i = 2**n
    y = f(x, i)

    ax1.plot(
        x, 
        y,
        linewidth=2.5,
        color=colors[idx],
        label=f'Literal = {i}'
    )
    
x_pow = 23
x_number = x**x_pow
y2 = getULP(x_number)

ax2.plot(
    x,
    y2,
    color="black",
    linewidth=2.5,
    label=r'ULP$(x^{23})$'
)

ax1.grid(True, linestyle="--", alpha=0.5)
ax2.grid(True, linestyle="--", alpha=0.5)

ax1.set_yscale("log", base=2)
ax2.set_yscale("log", base=2)

power_of_two_formatter = ticker.FuncFormatter(lambda y, _: r'$2^{{{:g}}}$'.format(np.log2(y)))
ax1.yaxis.set_major_formatter(power_of_two_formatter)
ax2.yaxis.set_major_formatter(power_of_two_formatter)

ax1.set_ylabel("f(x)")
ax2.set_ylabel("ULP")
ax2.set_xlabel("x")

ax1.legend(fontsize=20, loc=5)
ax2.legend(fontsize=20, loc=4)

plt.tight_layout()
plt.show()
````

````python
x = np.linspace(rangeX[0], rangeX[1], 1200, dtype=np.float64)

q = 6

fig3 = plt.figure(figsize=(inches_x*2, inches_y*2))

colors = plt.cm.viridis(np.linspace(0, 1, q))

for idx, n in enumerate(np.arange(0, q)):
    i = 2**n
    y = f(x, i)

    plt.plot(
        x, 
        y,
        linewidth=2.5,
        color=colors[idx],
        label=f'Literal = {i}'
    )

x_pow = 23
x_number = x**x_pow
y2 = getULP(x_number)

plt.plot(
    x,
    y2,
    color="black",
    linewidth=2.5,
    label=r'ULP$(x^{23})$'
)

plt.grid(True, linestyle="--", alpha=0.5)

plt.yscale("log", base=2)
plt.xlabel("x")

plt.legend(fontsize=16, loc=5)

plt.tight_layout()
plt.show()
````

````python
from matplotlib.gridspec import GridSpec

# 1. Escala Global de Tipografia
plt.rcParams.update({
    'font.size': 16,          # Fonte base maior
    'axes.labelsize': 18,     # Rótulos dos eixos (x, y)
    'xtick.labelsize': 14,    # Numeração do eixo X
    'ytick.labelsize': 14,    # Numeração do eixo Y
    'legend.fontsize': 14,    # Fonte da legenda
    'lines.linewidth': 2.5    # Espessura padrão das linhas
})

def getULP(x: np.ndarray) -> np.ndarray:
    x = np.asarray(x, dtype=np.float64)
    return np.nextafter(x, np.inf) - x

# Presumo que rangeX, inches_x, inches_y e f() já estejam instanciados
x = np.linspace(rangeX[0], rangeX[1], 1200, dtype=np.float64)
x_pow = 23
x_number = x**x_pow
y2 = getULP(x_number)

q = 7
colors = plt.cm.viridis(np.linspace(0, 1, q))
power_of_two_formatter = ticker.FuncFormatter(lambda val, _: r'$2^{{{:g}}}$'.format(np.log2(val)))

# 2. Redução do Comprimento (Eixo X da Figura)
# Reduzido de inches_x * 4 para * 2.8 para um aspecto mais vertical e compacto
fig = plt.figure(figsize=(inches_x * 2.8, inches_y * 2))

# wspace adiciona um respiro horizontal entre a coluna esquerda e direita
gs = GridSpec(nrows=2, ncols=2, figure=fig, wspace=0.1)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[1, 0], sharex=ax1)
ax3 = fig.add_subplot(gs[:, 1])

for idx, n in enumerate(np.arange(0, q)):
    i = 2**n
    y = f(x, i)
    
    # ax1 (Esquerda-Topo) e ax3 (Direita)
    ax1.plot(x, y, color=colors[idx], label=f'Literal = {i}')
    ax3.plot(x, y, color=colors[idx], label=f'Literal = {i}')

# Plot do ULP
ax2.plot(x, y2, color="black", label=r'ULP$(x^{23})$')
ax3.plot(x, y2, color="black", label=r'ULP$(x^{23})$')

# Configuração de Grid e Escalas
for ax in [ax1, ax2, ax3]:
    ax.grid(True, linestyle="--", alpha=0.5)
    if ax != ax3:
        ax.set_yscale("log", base=2)
        ax.yaxis.set_major_formatter(power_of_two_formatter)

ax3.set_yscale("log", base=2)
ax3.yaxis.set_major_formatter(power_of_two_formatter)

# Rótulos (Otimizados para legibilidade)
ax1.set_ylabel("f(x)")
ax2.set_ylabel("ULP")
ax2.set_xlabel("x")
ax3.set_xlabel("x")
ax3.set_ylabel("Amplitude") 

handles, labels = ax3.get_legend_handles_labels()

by_label = dict(zip(labels, handles))

fig.legend(
    by_label.values(), 
    by_label.keys(), 
           loc='upper center', 
           bbox_to_anchor=(0.5, 1.0),
           ncol=4,
           frameon=False,
        ) 

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````
