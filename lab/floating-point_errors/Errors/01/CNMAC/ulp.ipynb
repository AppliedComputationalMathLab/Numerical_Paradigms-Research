````python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import plotly.express as px
import pandas as pd 
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
import plotly.figure_factory as ff
import plotly.io as pio
import plotly.subplots as sp
import plotly.express as px
````

````python
def f(x: np.float64, literal: np.float64) -> np.float64:
    return (x**23) + literal - (x**23)
````

````python
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
````

````python
centimeters_x = 27
centimeters_y = 11.5

inches_x = centimeters_x / 2.54
inches_y = centimeters_y / 2.54


plt.gcf().set_size_inches(inches_x, inches_y)
````

````python
def getULP(x: np.float64) -> np.float64:
    return np.nextafter(x, np.inf) - x

fig1 = plt.figure()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.gcf().set_size_inches(18, 6)

rangeX = (4, 6)
x = np.linspace(rangeX[0], rangeX[1], 1200, dtype = np.float64)


x_pow = 23
x_number = x**x_pow

y1 = getULP(x_number)

plt.plot(x, y1, label=f'ULP of x^{x_pow}')

plt.legend(fontsize=12)
plt.xlim(rangeX)

plt.yscale("log", base=2)



fig1.show()
````
