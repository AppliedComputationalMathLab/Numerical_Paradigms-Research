```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import plotly.express as px
import pandas as pd 
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
import plotly.figure_factory as ff
import plotly.io as pio
import plotly.subplots as sp
import plotly.express as px
```

```python
def fx(x:  np.ndarray) -> np.ndarray:
    return (np.pow(x,3)/3) - 3*x + 3
def fPrime(x:  np.ndarray) -> np.ndarray:
    return np.pow(x,2) - 3
def NumericalDerivative(x: np.ndarray, h: np.float64) -> np.ndarray:
    return (fx(x + h) - fx(x - h)) / (2 * h)
def error(h: np.float64) -> np.ndarray:
    return np.abs(fPrime(1) - NumericalDerivative(1, h))
```

```python
rangeX = (0, 0.5e-3)
rangeY = (1e-15, 1e0)
```

```python
h = np.linspace(rangeX[0], rangeX[1], 4000, dtype=np.float64)
```

```python
plt.plot(h, error(h), label='Error', color='red', linewidth=1)
plt.xlabel('h')
plt.ylabel('Error')
plt.gcf().set_size_inches(18, 6)
plt.legend(fontsize=24)
plt.xlim(rangeX)
plt.ylim(rangeY)
plt.yscale('log')

plt.grid(True)
```
